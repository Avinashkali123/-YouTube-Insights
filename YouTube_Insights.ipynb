```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Convert all columns to numeric, coercing errors
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop columns that are all NaN (i.e., the original string columns)
df_numeric = df.dropna(axis=1, how='all')

# Create the correlation matrix
plt.figure(figsize=(20, 15))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of YouTube Analytics')
plt.show()
```

```python
# Create a scatter plot of views vs. likes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='likes', data=df_numeric)
plt.title('Views vs. Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

# Create a scatter plot of views vs. comments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='comments', data=df_numeric)
plt.title('Views vs. Comments')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.show()

# Create a scatter plot of views vs. subscribersGained
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='subscribersGained', data=df_numeric)
plt.title('Views vs. Subscribers Gained')
plt.xlabel('Views')
plt.ylabel('Subscribers Gained')
plt.show()
```

```python
# Convert the 'day' column to datetime
df_numeric['day'] = pd.to_datetime(df['day'])

# Plot the total views over time
plt.figure(figsize=(15, 7))
df_numeric.groupby('day')['views'].sum().plot()
plt.title('Total Views Over Time')
plt.xlabel('Date')
plt.ylabel('Total Views')
plt.show()
```

```python
# Reload the original data to get the 'day' column
df_original = pd.read_csv('all_youtube_analytics.csv')

# Merge the 'day' column into the numeric dataframe
df_numeric['day'] = df_original['day']

# Convert the 'day' column to datetime
df_numeric['day'] = pd.to_datetime(df_numeric['day'])

# Plot the total views over time
plt.figure(figsize=(15, 7))
df_numeric.groupby('day')['views'].sum().plot()
plt.title('Total Views Over Time')
plt.xlabel('Date')
plt.ylabel('Total Views')
plt.show()
```

```python
# Plot the total likes over time
plt.figure(figsize=(15, 7))
df_numeric.groupby('day')['likes'].sum().plot()
plt.title('Total Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.show()

# Plot the total comments over time
plt.figure(figsize=(15, 7))
df_numeric.groupby('day')['comments'].sum().plot()
plt.title('Total Comments Over Time')
plt.xlabel('Date')
plt.ylabel('Total Comments')
plt.show()

# Plot the total subscribers gained over time
plt.figure(figsize=(15, 7))
df_numeric.groupby('day')['subscribersGained'].sum().plot()
plt.title('Total Subscribers Gained Over Time')
plt.xlabel('Date')
plt.ylabel('Total Subscribers Gained')
plt.show()
```

```python
# Top 5 videos by views
top_videos_by_views = df_original.groupby('video_id')['views'].sum().nlargest(5)
print("Top 5 Videos by Views:")
print(top_videos_by_views)
print("\n")

# Top 5 videos by likes
top_videos_by_likes = df_original.groupby('video_id')['likes'].sum().nlargest(5)
print("Top 5 Videos by Likes:")
print(top_videos_by_likes)
print("\n")

# Top 5 videos by comments
top_videos_by_comments = df_original.groupby('video_id')['comments'].sum().nlargest(5)
print("Top 5 Videos by Comments:")
print(top_videos_by_comments)
```
